**We are using FB Prophet model**

converting the date format to yyyy-mm-dd

```python
date_info=pd.to_datetime(data['Date'],format='%d/%m/%Y') # Changed the format to Day/Month/Year
print(date_info)
```

```python
time_info=data['Time']
time_info=time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#combining 2 series to a pandas dataframe
data_time=pd.concat([date_info,time_info],axis=1)
data_time.head()
```

```python
#combining date and time
data_time['ds']=data_time['Date'].astype(str)+' '+data_time['Time'].astype(str)
data_time.head()
```

```python
data_time.info()
```

We have to convert ds from object datatype to date time datatype

```python
Data=pd.DataFrame()
```

```python
Data['ds']=pd.to_datetime(data_time['ds'])
```

```python
Data.head()
```

```python
Data['y']=data['AH']
```

```python
Data.head()
```

```python
!pip install prophet
```

```python
from prophet import Prophet
```

```python
model=Prophet()
model.fit(Data)
```

```python
future=model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

plot_plotly(model, forecast)
```
