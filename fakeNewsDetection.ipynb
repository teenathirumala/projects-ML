```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Try using the 'python' engine and handle bad lines
fake = pd.read_csv('Fake.csv', engine='python', on_bad_lines='skip')
```

```python
# Try using the 'python' engine and handle bad lines
true = pd.read_csv('True.csv', engine='python', on_bad_lines='skip')
```

```python
fake.head()
```

```python
true.head()
```

```python
true['label']=1
fake['label']=0
```

```python
news=pd.concat([fake,true],axis=0)
news.head()
```

```python
news.tail()
```

```python
news.info()
```

```python
news.dropna()
```

```python
news.drop(['title','subject','date'],axis=1,inplace=True)
news.head()
```

```python
news.isnull().sum()
```

```python
#we shd reshuffle it as there migth be chance of overfitting as the data looks so arranged
news=news.sample(frac=1)
news.head()
```

```python
news.reset_index(inplace=True)
news.head()
```

```python
news.drop(['index'],axis=1,inplace=True)
news.head()
```

```python
import re
```

```python
def wordopt(text):
    if text is not None: # Check if text is not None
        text=text.lower()
        text=re.sub('\[.*?\]','',text)
        text=re.sub('https?://\S+|www\.\S+','',text)
        text=re.sub('<.*?>+','',text)
        text=re.sub('[%s]','',text)
        text=re.sub('\n','',text)
        text=re.sub('\w*\d\w*','',text)

        return text
    else:
        return "" # Return empty string if text is None
```

```python
news['text']=news['text'].apply(wordopt)
news.head()
```

```python
x=news['text']
y=news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
vec_train=vec.fit_transform(x_train)
vec_test=vec.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(vec_train,y_train)
```

```python
y_pred=log.predict(vec_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d',cmap='Blues')
plt.show()
```

```python
#same problem but with decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(vec_train,y_train)
```

```python
dt_pred=dt.predict(vec_test)
dt_pred
```

```python
dt.score(vec_test,y_test)
```

```python
print(classification_report(y_test,dt_pred))
```

```python
#random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(vec_train,y_train)
```

```python
rf_pred=rf.predict(vec_test)
rf_pred
```

```python
rf.score(vec_test,y_test)
```

```python
print(classification_report(y_test,rf_pred))
```

```python
#with gbc
from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier()
gbc.fit(vec_train,y_train)
```

```python
gbc_pred=gbc.predict(vec_test)
gbc_pred
```

```python
gbc.score(vec_test,y_test)
```

```python
print(classification_report(y_test,gbc_pred))
```

```python
def output_label(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "True News"
```

```python
def manual_testing(news):
  testing_news={"text":[news]}
  new_def_test=pd.DataFrame(testing_news)
  new_def_test
  new_def_test["text"]=new_def_test["text"].apply(wordopt)
  new_x_test=new_def_test["text"]
  new_vec_test=vec.transform(new_x_test)
  pred_gbc=gbc.predict(new_vec_test)
  return output_label(pred_gbc[0])
```

```python
news_article=str(input())
manual_testing(news_article)
```
