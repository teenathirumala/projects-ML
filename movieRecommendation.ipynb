```python
data['tags']=data['genre']+data['overview']
```

```python
data.head()
```

```python
new_df=data[['id','title','tags']]
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
vec=cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
vec
```

```python
vec.shape
```

```python
#we are finding te similaruty between movies
from sklearn.metrics.pairwise import cosine_similarity
sim=cosine_similarity(vec)
sim
```

```python
dist=sorted(list(enumerate(sim[0])),reverse=True,key=lambda x:x[1])
dist
```

```python
for i in dist[0:5]:
    print(new_df.iloc[i[0]].title)
```

```python
def recommend(movie):
     #try:
        index = new_df[new_df['title']==movie].index[0]
        dist = sorted(list(enumerate(sim[index])),reverse=True,key=lambda vec:vec[1])
        for i in dist[0:5]:
            print(new_df.iloc[i[0]].title)
   # except IndexError:
        #print(f"Movie '{movie}' not found in the dataset.")
```

```python
recommend('Avatar')
```
